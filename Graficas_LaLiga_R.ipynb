```r
library(ggplot2)
library(tidyr)
library(dplyr)
```

```r
data_equipos <- read.csv('df_equipos.txt', sep = '\t', header = TRUE)
```

```r
data_jugadores <- read.csv('df_jugadores.txt',sep = '\t', header = TRUE)
```

```r
# Goles
ggplot(data_equipos, mapping = aes (x = Equipo, y = Gls.)) + geom_col(fill = "blue", color = "black") +
ylab("Goles") + theme(axis.text.x = element_text(angle = 75, hjust = 1)) +
scale_y_continuous(breaks = seq(0, max(data_equipos$Gls.), by = 5))
```

```r
ggsave("Goles_por_equipo.png", width = 5, height = 5)
```

```r
# Asistencias
ggplot(data_equipos, mapping = aes (x = Equipo, y = Ass)) + geom_col(fill = "blue", color = "black") +
ylab("Asistencias") + theme(axis.text.x = element_text(angle = 75, hjust = 1))+
scale_y_continuous(breaks = seq(0, max(data_equipos$Ass), by = 5))
ggsave("Asistencias_por_equipo.png", width = 5, height = 5)
```

```r
# Goles y asistencias, para ver qué equipos marcan más goles a través de jugada colectiva (por asistencia)
# Los que más lo harán serán en los que menos diferencia haya entre barra azul oscura y barra azul clarita

ggplot(data_equipos, aes(x = Equipo)) +
    geom_bar(aes(y = Gls., fill = "Goles"), stat = "identity") +
    geom_bar(aes(y = Ass, fill = "Asistencias"), stat = "identity") +
    ylab("Goles y Asistencias") +
    scale_fill_manual(values = c("Goles" = "blue", "Asistencias" = "lightblue")) +
    theme(axis.text.x = element_text(angle = 75, hjust = 1))
ggsave("GolesyAsistencias_por_equipo.png", width = 5, height = 5)
```

```r
# Posesión media

ggplot(data_equipos, aes(x = Equipo, y = Pos.)) +
    geom_col(fill = "blue", color = "black") +
    ylab("Posesión (%)") +
    theme(axis.text.x = element_text(angle = 75, hjust = 1)) +
    scale_y_continuous(breaks = seq(0, 100, by = 10), limits = c(0, 100))
ggsave("Posesion_por_equipo.png", width = 5, height = 5)
```

```r
# # a ver si los equipos con mas posesion generan mas goles+asistencias(mas jugadas de gol vamos)
ggplot(data_equipos, aes(x = Equipo, y = G.A, fill = Pos.)) +
    geom_bar(stat = "identity") +
    scale_fill_gradient(low = "lightblue", high = "darkblue") +
    ylab("Goles+Asistencias") +
    theme(axis.text.x = element_text(angle = 75, hjust = 1)) 
ggsave("Goles+Asistencias_Vs_Posesion_por_equipo.png", width = 5, height = 5)
```

```r
# Tarjetas amarillas

ggplot(data_equipos, mapping = aes (x = Equipo, y = TA)) + geom_col(fill = "blue", color = "black") +
ylab("Tarjetas Amarillas") + theme(axis.text.x = element_text(angle = 75, hjust = 1))+
scale_y_continuous(breaks = seq(0, max(data_equipos$TA), by = 5))

ggsave("TA_por_equipo.png", width = 5, height = 5)
```

```r
# Tarjetas rojas

ggplot(data_equipos, mapping = aes (x = Equipo, y = TR)) + geom_col(fill = "blue", color = "black") +
ylab("Tarjetas Rojas") + theme(axis.text.x = element_text(angle = 75, hjust = 1))+
scale_y_continuous(breaks = seq(0, max(data_equipos$TR), by = 1))
ggsave("TR_por_equipo.png", width = 5, height = 5)
```

```r
# Ambas, tarjetas amarillas y rojas y el total
data_equipos$Total <- data_equipos$TA + data_equipos$TR

# Reorganizamos el dataframe en formato largo (long format) usando tidyr
data_long <- pivot_longer(data_equipos, cols = c(TA, TR, Total), names_to = "Variable", values_to = "Value")

ggplot(data_long, aes(x = Equipo, y = Value, fill = Variable)) +
    geom_bar(stat = "identity", position = "dodge") +
    ylab("Tarjetas") +
    xlab("Equipo") +
    scale_fill_manual(values = c("TA" = "yellow2", "TR" = "red3", "Total" = "blue3")) +
    theme(axis.text.x = element_text(angle = 75, hjust = 1)) +
    scale_y_continuous(breaks = seq(0, max(data_equipos$Total), by = 5))


ggsave("Tarjetas_Totales_por_equipo.png", width = 5, height = 5)
# En este código, hemos reorganizado el dataframe a un formato largo utilizando 
# pivot_longer, lo que nos permite tener una columna "variable" que distingue entre tarjetas amarillas ("TA"), tarjetas rojas ("TR") y el total ("Total"). Luego, hemos utilizado geom_bar para crear un gráfico de barras con las tres barras para cada equipo, una para cada tipo de tarjeta.
```

```r
# Edad equipos
ggplot(data_equipos, mapping = aes (x = Equipo, y = Edad)) + geom_col(fill = "blue", color = "black") +
ylab("Edad Media") + theme(axis.text.x = element_text(angle = 75, hjust = 1))+
scale_y_continuous(breaks = seq(0, max(data_equipos$Edad), by = 5))
ggsave("Edad_media_por_equipo.png", width = 5, height = 5)
```

```r
# Vamos a hacer lo mismo con goles y asistencias

# Reorganizar el dataframe en formato largo (long format) usando tidyr

data_long <- pivot_longer(data_equipos, cols = c(Gls., Ass, G.A), names_to = "Variable", values_to = "Value")

ggplot(data_long, aes(x = Equipo, y = Value, fill = Variable)) +
  geom_bar(stat = "identity", position = "dodge") +
  ylab("Valores") +
  xlab("Equipo") +
  scale_fill_manual(values = c("Gls." = "green3", "Ass" = "cyan3", "G.A" = "blue3")) +
  theme(axis.text.x = element_text(angle = 75, hjust = 1))
ggsave("Goles+Asistencias_por_equipo.png", width = 5, height = 5)
```

```r
# Vamos a analizar los jugadores

# Máximos goleadores

# Ordenamos el dataframe por la columna "Gls." en orden descendente
goleadores_ordenados <- data_jugadores %>%
  arrange(desc(Gls.))

# Seleccionar los 10 máximos goleadores (puedes ajustar el número)
top_goleadores <- head(goleadores_ordenados, 20)

# Crear un gráfico de barras con los nombres de los jugadores en el eje x y los goles en el eje y
ggplot(top_goleadores, aes(x = Jugador, y = Gls.)) +
  geom_bar(stat = "identity", fill = "blue") +
  ylab("Goles") +
  xlab("Jugador") +
  theme(axis.text.x = element_text(angle = 75, hjust = 1))

ggsave("Goles_por_jugador.png", width = 5, height = 5)
```

```r
# Maximos goleadores ordenados
top_goleadores <- head(goleadores_ordenados, 20)
ggplot(top_goleadores, aes(x = reorder(Jugador, -Gls.), y = Gls.)) +
geom_bar(stat = "identity", fill = "blue", color = "black") +
ylab("Goles") +
xlab("Jugador") +
theme(axis.text.x = element_text(angle = 75, hjust = 1)) +
scale_y_continuous(breaks = seq(0, max(top_goleadores$Gls.), by = 1))
ggsave("Goles_por_jugador_ordenados.png", width = 5, height = 5)
```
